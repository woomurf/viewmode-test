What if Van Gogh had painted the Startled Cat? 
What would it look like? Let's use style transfer and try to build a possible replica
![Imgur](https://i.imgur.com/ImEN6MV.png)

Recommended Reading: [Style Transfer CVPR Paper](https://www.cv-foundation.org/openaccess/content_cvpr_2016/papers/Gatys_Image_Style_Transfer_CVPR_2016_paper.pdf)

```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

from PIL import Image
import numpy as np

import torch
from torch import optim
from torchvision import transforms, models
```

```python
# 변경된다!!
```

# Download model for Style Transfer

```python
vgg = models.vgg19(pretrained=True)
vgg_features = vgg.features
```

```python
# Turn off gradient update
for param in vgg_features.parameters():
    param.requires_grad = False
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
vgg_features.to(device)
```

# Load data

```python
def fetch_img(path, shape=None, size=400, debug=False):
    """
    Function to load in an image and transform it to a tensor
    
    Args:
        path: path of the input image
        shape: explicit shape (if needed)
        size: set the size of image tensor
        debug: If true, display image

    Returns:
        image tensor on available device
    """

    img = mpimg.imread(path)

    if debug:
        plt.imshow(img)

    if max(img.shape) < size:
        size = max(img.size)

    # if an explicit shape is passed, then override size to that shape
    if shape:
        size = shape

    transform = transforms.Compose([transforms.ToPILImage(),
                                    transforms.Resize(size), transforms.ToTensor(), 
                                    transforms.Normalize((0.485, 0.456, 0.406), 
                                                        (0.229, 0.224, 0.225))])
    # unsqueeze(0) adds a batch dimension
    img_tensor = transform(img)[:3,:,:].unsqueeze(0)
    return img_tensor.to(device)
```

```python
style_path = "starry_night.jpg" # specify the path of your own style image
style_tensor = fetch_img(style_path, debug=True)
```

```python
content_path = "TheCat.jpg" # specify the path of the image on which you want to perform style transfer
content_tensor = fetch_img(content_path, shape=style_tensor.shape[-2:], debug=True)
```

# Extract features for style transfer

Based on the [StyleGAN CVPR Paper](https://www.cv-foundation.org/openaccess/content_cvpr_2016/papers/Gatys_Image_Style_Transfer_CVPR_2016_paper.pdf) let's define a function to fetch the feature representations from Convolution layers.

```python
def fetch_conv_feat(model, img, layers=None):

    conv_features = {}

    # if layers are not provided, use default layers (based on StyleGAN paper)
    if not layers:
        layers = {'0': 'conv1_1',
                  '5': 'conv2_1', 
                  '10': 'conv3_1', 
                  '19': 'conv4_1',
                  '21': 'conv4_2',
                  '28': 'conv5_1'}

    for name, layer in model._modules.items():
        img = layer(img)
        if name in layers:
            conv_features[layers[name]] = img

    return conv_features
```

```python
def tensor_to_img(tensor):
    """
    Convert the input tensor back to image

    Args: 
        tensor: input image tensor

    Returns:
        image in numpy array
    """

    tensor = tensor.squeeze().permute(1, 2, 0)  # remove batch dim and swap axes for numpy
    tensor = tensor.to("cpu").clone().detach()  # move tensor to cpu 
    img = tensor.numpy()
    img = img * np.array((0.229, 0.224, 0.225)) + np.array((0.485, 0.456, 0.406))  # undo normalisation
    img = img.clip(0, 1)

    return img
```

# Calculate Gram Matrix
Next, based on the paper we'll define a function to calculate the gram matrix of a given tensor.  
Steps to calculate gram matrix are as follows:
1) Flatten the spatial dimensions of the input tensor
2) Multiply the reshaped tensor by its transpose to get the gram matrix

Reference: [Wikipedia](https://en.wikipedia.org/wiki/Gramian_matrix)

```python
def gram_matrix(tensor):
    
    batch_size, depth, height, width = tensor.size()
    
    # flatten the spatial data
    tensor = tensor.view(depth, height * width)

    gram_mat = torch.mm(tensor, tensor.t())
    
    return gram_mat 
```

```python
style_features = fetch_conv_feat(vgg_features, style_tensor)
content_features = fetch_conv_feat(vgg_features, content_tensor)

style_grams = {layer: gram_matrix(style_features[layer]) for layer in style_features}

# create a clone of the transfer image tensor
# this tensor will be changed iteratively 
target_tensor = content_tensor.clone().requires_grad_(True).to(device)
```

# Layer Weighting
We need to supply the weighting for the layers to decide the style influence of each layer in the transformation.

```python
style_weights = {'conv1_1': 1.,
                 'conv2_1': 0.75,
                 'conv3_1': 0.2,
                 'conv4_1': 0.2,
                 'conv5_1': 0.2}
```

```python
# define an alpha (content_weight) and a beta (style_weight) based on the StyleGAN paper
content_weight = 1  # alpha
style_weight = 1e6  # beta
```

# Style Transfer
Content and Style losses  
1) Content loss - mean squared difference between the target and content features at layer conv4_2.  
   `content_loss = torch.mean((target_features['conv4_2'] - content_features['conv4_2'])**2)`  

2) Style loss - calculated in a similar way. Just iterate through the layers specified in the dictionary `style_weights`.

```python
display_rate = 400

optimizer = optim.Adam([target_tensor], lr=0.001)
steps = 8000

for idx in range(1, steps + 1):
    
    # fetch features from target image
    target_features = fetch_conv_feat(vgg_features, target_tensor)

    content_loss = torch.mean((target_features['conv4_2'] - content_features['conv4_2'])**2)

    style_loss = 0

    for layer in style_weights:
        target_feature = target_features[layer]
        target_gram = gram_matrix(target_feature)
        batch_size, depth, height, width = target_feature.shape

        style_gram = style_grams[layer]

        layer_style_loss = style_weights[layer] * torch.mean((target_gram - style_gram)**2)

        style_loss += layer_style_loss / (depth * height * width)
        
    total_loss = content_weight * content_loss + style_weight * style_loss
    
    # update target image
    optimizer.zero_grad()
    total_loss.backward()
    optimizer.step()
    
    # print loss and display intermediate images
    if  idx % display_rate == 0:
        print('Total loss: ', total_loss.item())
        plt.imshow(tensor_to_img(target_tensor))
        plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 10))
ax1.imshow(tensor_to_img(content_tensor))
ax2.imshow(tensor_to_img(target_tensor))
```

```python
mpimg.imsave("starry_cat.jpg", tensor_to_img(target_tensor))
```
